```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = data.drop(data.index[-1])
df
```

```python
df.describe()
```

```python
df.plot(x='Producto', y=['Ventas'], kind='line', marker='o')
plt.title('Ventas de Productos')
plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.show()
```

```python
df.plot(x='Vendedor', y=['Ventas'], kind='bar',color =['#FF5733', '#33C3FF'])
plt.title('Ventas Totales')
plt.xlabel('Vendedor')
plt.ylabel('Ventas')
plt.show()
```

```python
df[['Producto','Cantidad', 'Precio', 'Ventas']]
```

```python
vendedor_especifico = 'Nancy Gil de la Peña'
ventas_vendedor = df[df['Vendedor'] == vendedor_especifico]
print(ventas_vendedor)
```

```python
total_ventas = ventas_vendedor['Cantidad'] * ventas_vendedor['Precio'] # Changed 'cantidad' to 'Cantidad'
total_ventas = total_ventas.sum()
print(f"El total de ventas de {vendedor_especifico} es: {total_ventas}")
```
