```python
df = data.drop(data.index[-1])
df
```

```python
df['Percentage_of_iPhone_Users'] = df['Percentage_of_iPhone_Users'].ffill()
df['No_of_iPhone_Sold_USA'] = df['No_of_iPhone_Sold_USA'].bfill()
df['iOS_Market_Share'] = df['iOS_Market_Share'].ffill()
df['Android_Market_Share'] = df['Android_Market_Share'].ffill()
```

```python
df.isna().sum()
```

```python
df[['Year','No_of_iPhone_Users','No_of_iPhone_Users_USA']]
```

```python
df.plot(x='Year', y=['No_of_iPhone_Users', 'No_of_iPhone_Users_USA'], kind='line', marker='o')
plt.title('Trend of iPhone Users Globally and in the USA (2011-2023)')
plt.ylabel('Number of Users')
plt.show()
```

```python
df[['Year','No_of_iPhone_Sold','No_of_iPhone_Sold_USA']]
```

```python
df.plot(x='Year', y=['No_of_iPhone_Sold', 'No_of_iPhone_Sold_USA'], kind='line',marker='o')
plt.title('Trend of iPhone Sales Globally and in the USA (2011-2023)')
plt.ylabel('Number of iPhones Sold')
plt.show()
```

```python
df[['Year','iOS_Market_Share','Android_Market_Share']]
```

```python
df.plot(x='Year', y=['iOS_Market_Share', 'Android_Market_Share'], kind='bar',color =['#FF5733', '#33C3FF']  )
plt.title('iOS vs Android Market Share (2011-2023)')
plt.ylabel('Market Share (%)')
plt.show()
```

```python
df.describe()
```

```python
df['USA_Percentage_of_Global_Users'] = (df['No_of_iPhone_Users_USA'] / df['No_of_iPhone_Users']) * 100
df[['Year', 'USA_Percentage_of_Global_Users']]
```

```python
df.plot(x='Year', y='USA_Percentage_of_Global_Users', kind='bar')
plt.title('Percentage of Global Users in the USA (2011-2023)')
plt.ylabel('Percentage (%)')
plt.show()
```

```python
df['differ_df']=df['No_of_iPhone_Users']-df['iOS_Market_Share']
df[['Year','differ_df']]
```

```python
plt.figure(figsize=(8, 6))
plt.pie(df['No_of_iPhone_Users'], labels=df['Year'], autopct='%1.1f%%', startangle=140,colors=['#ff9999', '#ff66b3', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c2f0c2'])
plt.title('Distribution of iPhone Users Over the Years')
plt.show()
```
